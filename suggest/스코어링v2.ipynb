```python
import requests

# 유저 ID와 API URL 설정
user_id = 1
api_url = f"http://107.23.187.64:8000/api/like-accommodation/?user_id={user_id}"

# 좋아요 정보를 받아오는 함수
def fetch_liked_hotels(api_url):
    try:
        # API GET 요청
        response = requests.get(api_url)

        # 응답 상태 확인
        if response.status_code == 200:
            # JSON 데이터를 파싱하여 좋아요 호텔 리스트 반환
            liked_hotels = response.json()
            print(f"사용자 {user_id}의 좋아요 호텔 데이터: {liked_hotels}")
            return liked_hotels
        else:
            # 요청 실패 시 상태 코드와 메시지 출력
            print(f"요청 실패: 상태 코드 {response.status_code}, 내용: {response.text}")
            return []
    except Exception as e:
        # 예외 처리
        print(f"요청 중 오류 발생: {e}")
        return []

# 좋아요 호텔 데이터 가져오기
liked_hotels = fetch_liked_hotels(api_url)

# 가져온 데이터를 활용
if liked_hotels:
    hotel_names = [hotel["name"] for hotel in liked_hotels]  # 호텔 이름 추출
    print(f"사용자가 좋아요를 누른 호텔 이름 목록: {hotel_names}")
else:
    print("사용자가 좋아요를 누른 호텔이 없습니다.")
```
