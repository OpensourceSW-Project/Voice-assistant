```python
"""N/A값 들어간 데이터 다시 크롤링하는 코드"""

# 크롬 드라이버 경로 설정
chrome_driver_path = 'c:/Users/yeonj/OneDrive/바탕 화면/data_crawling/chromedriver.exe'

# 단일 URL
test_url = 'https://place-site.yanolja.com/places/3001918?checkinDate=2024-11-19&checkoutDate=2024-11-20&capacityAdult=2&capacityChild=0'

def yanolja_crawl(url):
    # 드라이버 시작
    service = Service(chrome_driver_path)
    service.start()
    driver = webdriver.Chrome(service=service)
    driver.get(url)
    df = pd.DataFrame(columns=['hotel', 'star', 'review', 'date', 'price', 'address'])

    # 호텔 이름 가져오기
    try:
        hotel_element = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'h1.css-qnrj3c'))
        )
        hotel = hotel_element.text
    except Exception as e:
        print(f"Error while finding hotel: {e}")
        hotel = "N/A"

    # 주소지 버튼 클릭 및 주소 복사
    try:
        address_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/div/main/article/div[1]/div/div[1]/div[3]/div/div'))
        )
        address_button.click()
        time.sleep(2)

        copy_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="_MODAL_DIM_"]/div/div/div/main/div/div[2]/div[2]/div[2]/div/span'))
        )
        copy_button.click()
        time.sleep(1)
        address = pyperclip.paste()

        close_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="_MODAL_DIM_"]/div/div/div/header/div/div[1]/img'))
        )
        close_button.click()
        time.sleep(1)
    except Exception as e:
        print(f"Error while copying address: {e}")
        address = "N/A"

    # 최신 상단 가격 가져오기
    try:
        driver.execute_script("window.scrollBy(0, 800);")
        time.sleep(3)
        price_element = WebDriverWait(driver, 40).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'span.price'))
        )
        price = price_element.text
    except Exception as e:
        print(f"Error while finding price: {e}")
        price = "N/A"

    # 리뷰 페이지 이동 및 리뷰 수집
    try:
        review_button = WebDriverWait(driver, 15).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/div/main/article/div[1]/div/div[1]/div[4]/a/span'))
        )
        driver.execute_script("arguments[0].click();", review_button)
        time.sleep(2)

        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1.5)
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")
            time.sleep(1.5)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        review_containers = driver.find_elements(By.CSS_SELECTOR, "div.review-item-container")
        for container in review_containers:
            try:
                review_text_element = container.find_element(By.CSS_SELECTOR, "p.content-text.css-c92dc4")
                review_text = review_text_element.text
            except Exception as e:
                print(f"Skipping a review due to missing or hidden content: {e}")
                continue

            try:
                star_container = container.find_element(By.CSS_SELECTOR, "div.css-rz7kwu")
                stars = star_container.find_elements(By.TAG_NAME, "svg")
                star_count = 0
                for star in stars:
                    paths = star.find_elements(By.TAG_NAME, "path")
                    fill_values = [path.get_attribute("fill") for path in paths]
                    if "#FDBD00" in fill_values:
                        star_count += 1
            except Exception as e:
                print(f"Error while extracting stars: {e}")
                star_count = 0

            try:
                date_element = container.find_element(By.CSS_SELECTOR, "p.css-1irbwe1")
                date_text = date_element.text
            except Exception as e:
                print(f"Error while extracting date: {e}")
                date_text = "N/A"

            df.loc[len(df)] = [hotel, star_count, review_text, date_text, price, address]
    except Exception as e:
        print(f"Error while extracting reviews: {e}")

    # 데이터프레임을 CSV 파일로 저장
    output_dir = './review'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    output_filename = os.path.join(output_dir, 'yanolja_data_57.csv')
    df.to_csv(output_filename, index=False, encoding='utf-8')
    print(f"Data saved to {output_filename}")

    # 드라이버 종료
    driver.quit()
    service.stop()

# 단일 URL 크롤링
yanolja_crawl(test_url)


```
